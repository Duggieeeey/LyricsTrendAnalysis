```python
#1. Data Collecting

import pandas as pd
import lyricsgenius

file_path = 'C:/Users/withm/OneDrive/Desktop/Lyrics Project/playlist_2010to2022.csv'
df = pd.read_csv(file_path)

# Extract the relevant columns
selected_columns = df[['year', 'track_name', 'artist_name']]

# Create a list of years from 2000 to 2022
years = list(range(2000, 2023))
```

```python
genius = lyricsgenius.Genius('AKyhQ-Q_PyaEJwKOGjEpmei3_SL5MBIqm99Dlzf7XgQJBgxKP8c8l8Nh-QP8rHS5', timeout=10)
```

```python
# Loop through each year, search for lyrics, and append them to a text file
for year in years:
    # Filter the DataFrame for songs from the current year
    songs_for_year = selected_columns[selected_columns['year'] == year]

    # Create a text file for the current year
    file_name = f'lyrics_{year}.txt'
    
    with open(file_name, 'w', encoding='utf-8') as file:
        for index, row in songs_for_year.iterrows():
            # Search for the lyrics of the song
            song = genius.search_song(row['track_name'], row['artist_name'])
            
            if song:
                # Write the lyrics to the text file
                file.write(f"Lyrics:\n{song.lyrics}\n\n")
            else:
                # If lyrics are not found, print a message to the console
                print(f"Lyrics not found for '{row['track_name']}' by '{row['artist_name']}' in {year}")
```

```python
#2. Data Cleaning

import os
import re
import nltk
from nltk.corpus import stopwords

# Directory containing the lyrics text files
lyrics_directory = r'C:\Users\withm\OneDrive\Desktop\Lyrics Project\Cleaned Data'

# Define the words to remove and the pattern for text enclosed in square brackets
words_to_remove = ["lyrics", "hey", "uh-huh", "woo", "yeah", "ooh-ooh", "oh", "ayy", "nt", "s", "m", "re", "scp", 
                   "feat", "lala"]
pattern = r'\[.*?\]'

# Characters to remove
characters_to_remove = r'[,"\'!-?,—)–:(`.]'

# Get the list of NLTK English stop words
stop_words = set(stopwords.words("english"))

# Iterate through all the text files in the directory
for filename in os.listdir(lyrics_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(lyrics_directory, filename)

        # Open the file for reading
        with open(file_path, "r", encoding="utf-8") as file:
            lyrics_text = file.read()

        # Convert to lowercase
        lyrics_text = lyrics_text.lower()

        # Remove specified words
        for word in words_to_remove:
            # Use re.sub to remove word regardless of case (case-insensitive)
            lyrics_text = re.sub(rf'\b{re.escape(word)}\b', '', lyrics_text, flags=re.IGNORECASE)
            
        # Tokenize the text into words
        words = nltk.word_tokenize(lyrics_text)

        # Filter out stopwords
        filtered_words = [word for word in words if word not in stop_words]

        # Join the filtered words to form the cleaned lyrics
        cleaned_lyrics = ' '.join(filtered_words)

        # Remove text enclosed in square brackets
        cleaned_lyrics = re.sub(pattern, '', cleaned_lyrics)

        # Remove specified characters
        cleaned_lyrics = re.sub(characters_to_remove, '', cleaned_lyrics)

        # Save the cleaned text back to the file
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(cleaned_lyrics)
```

```python
#3. Feature Extraction
#(1) Word Frequencies

import os
from collections import Counter

# Directory containing tokenized lyrics text files for each year
lyrics_directory = r'C:\Users\withm\OneDrive\Desktop\Lyrics Project\Cleaned Data'

# Initialize Counters for each year
yearly_counters = {str(year): Counter() for year in range(2000, 2024)}

# Iterate through text files for each year
for year in range(2000, 2023):
    file_name = f'lyrics_{year}.txt'
    file_path = os.path.join(lyrics_directory, file_name)
    
    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            # Tokenized lyrics as a list of words
            tokenized_lyrics = file.read().split()  # Assuming words are space-separated
            # Update the corresponding year's Counter with word frequencies
            yearly_counters[str(year)].update(tokenized_lyrics)

# Print the 10 most common words for each year
for year, word_counts in yearly_counters.items():
    most_common_words = word_counts.most_common(10)
    print(f"Year {year}:")
    for word, frequency in most_common_words:
        print(f"{word}: {frequency}")
    print()  # Add a newline between years for clarity
```

```python
#(2) Sentimental Analysis

import os
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Directory containing tokenized lyrics text files for each year
lyrics_directory = r'C:\Users\withm\OneDrive\Desktop\Lyrics Project\Cleaned Data'

# Initialize a dictionary to store sentiment scores for each year
sentiment_scores = {str(year): {'positive': 0, 'neutral': 0, 'negative': 0} for year in range(2000, 2024)}

# Iterate through text files for each year
for year in range(2000, 2023):
    file_name = f'lyrics_{year}.txt'
    file_path = os.path.join(lyrics_directory, file_name)
    
    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            # Tokenized lyrics as a list of words
            tokenized_lyrics = file.read().split()
            # Calculate sentiment scores for the lyrics
            sentiment = sia.polarity_scores(' '.join(tokenized_lyrics))
            # Update sentiment scores for the corresponding year
            sentiment_scores[str(year)]['positive'] += sentiment['pos']
            sentiment_scores[str(year)]['neutral'] += sentiment['neu']
            sentiment_scores[str(year)]['negative'] += sentiment['neg']
            
            # Print sentiment scores for the current year
            print(f"Sentiment Scores for {year}:")
            print(f"Positive: {sentiment['pos']:.3f}")
            print(f"Neutral: {sentiment['neu']:.3f}")
            print(f"Negative: {sentiment['neg']:.3f}")
            print()  # Add a newline for clarity
```

```python
#(3-1) Extract bi-grams

import os
from collections import Counter
import nltk

# Directory containing tokenized lyrics text files for each year
lyrics_directory = r'C:\Users\withm\OneDrive\Desktop\Lyrics Project\Cleaned Data'

# Initialize a dictionary to store bigrams for each year
ngram_size = 2 
yearly_ngrams = {str(year): Counter() for year in range(2000, 2023)}

# Iterate through all years from 2000 to 2023
for year in range(2000, 2023):
    # Construct the file name for the current year
    file_name = f'lyrics_{year}.txt'
    file_path = os.path.join(lyrics_directory, file_name)

    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            # Tokenized lyrics as a list of words
            tokenized_lyrics = file.read().split()
            # Extract bigrams using Python's zip function
            bigrams = list(zip(tokenized_lyrics, tokenized_lyrics[1:]))
            # Count and store bigrams for the corresponding year
            yearly_ngrams[str(year)].update(bigrams)

# Print the most common bigrams for each year
for year in range(2000, 2023):
    year_str = str(year)
    most_common_ngrams = yearly_ngrams[year_str].most_common(10)
    print(f"Most common {ngram_size}-grams in {year_str}:")
    for ngram, frequency in most_common_ngrams:
        print(f"{' '.join(ngram)}: {frequency}")
    print()  # Add a newline for clarity between years
```

```python
#(3-2 Extract tri-grams

# Initialize a dictionary to store trigrams for each year
ngram_size = 3 
yearly_trigrams = {str(year): Counter() for year in range(2000, 2023)}

# Iterate through all years from 2000 to 2023
for year in range(2000, 2023):
    # Construct the file name for the current year
    file_name = f'lyrics_{year}.txt'
    file_path = os.path.join(lyrics_directory, file_name)

    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            # Tokenized lyrics as a list of words
            tokenized_lyrics = file.read().split()  # Assuming words are space-separated
            # Extract trigrams using Python's zip function
            trigrams = [tuple(tokenized_lyrics[i:i + 3]) for i in range(len(tokenized_lyrics) - 2)]
            # Count and store trigrams for the corresponding year
            yearly_trigrams[str(year)].update(trigrams)

# Print the most common trigrams for each year
for year in range(2000, 2023):
    year_str = str(year)
    most_common_trigrams = yearly_trigrams[year_str].most_common(10)
    print(f"Most common {ngram_size}-grams in {year_str}:")
    for trigram, frequency in most_common_trigrams:
        print(f"{' '.join(trigram)}: {frequency}")
    print()  # Add a newline for clarity between years
```
